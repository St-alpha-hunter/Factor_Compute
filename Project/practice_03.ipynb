```python
def build_row(security_id, data):
    row = {
        "SecurityID": security_id,
        "VOLUME_RATIO": data.get("volume_ratio"),

        # ACT
        "ACT": data.get("act_ratio", {}).get("overall"),
        "L_ACT": data.get("act_ratio", {}).get("by_size", {}).get("L"),
        "M_ACT": data.get("act_ratio", {}).get("by_size", {}).get("M"),
        "S_ACT": data.get("act_ratio", {}).get("by_size", {}).get("S"),
        "X_ACT": data.get("act_ratio", {}).get("by_size", {}).get("X"),

        # ORDER_RATIO
        "XLS_ORDER_RATIO": data.get("order_ratio", {}).get("by_size", {}).get("X"),
        "L_ORDER_RATIO": data.get("order_ratio", {}).get("by_size", {}).get("L"),
        "M_ORDER_RATIO": data.get("order_ratio", {}).get("by_size", {}).get("M"),
        "S_ORDER_RATIO": data.get("order_ratio", {}).get("by_size", {}).get("S"),
    }
    return row
```

```python
rows = []

for security_id, data in results.items():
    row = build_row(security_id, data)
    rows.append(row)
```

```python
df = pd.DataFrame(rows)

df.to_csv(
    "factorValue_YF.csv",
    index=False,
    encoding="utf-8-sig"
)
```
